Aggregate functions allow us to summarize tables with statistics like `mean`, `minimum`, and `maximum`. These provide insights into our database. 

The difference between `COUNT(*)` and `COUNT(column)` is that 
* all the table's rows are counted in the first case. 
* In the second case, only the non-null values of the specified column are counted.

### 4. Multiple Group Summary Statistics

When we're using only one criterion, there are as many groups as there are distinct values in the grouping column. When we're using more than one criterion, it's the combination of all the grouping columns that must be distinct.
